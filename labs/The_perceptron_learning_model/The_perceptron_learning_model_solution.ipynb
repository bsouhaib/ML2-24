# <font color='blue'>Exercise 1</font>
Prove that the PLA eventually converges to a linear separator for separable data. The following steps will guide you through the proof. Let $\mathbf{w}^*$ be an optimal set of weights (one which separates the data). The essential idea in this proof is to show that the PLA weights $\mathbf{w}(t)$ get "more aligned" with $\mathbf{w}^*$ with every iteration. For simplicity, assume that $\mathbf{w}(0) = 0$.

(a) Let $\rho = \min_{1 \leq n \leq N} y_n ({\mathbf{w}^*}^T \mathbf{x}_n)$. Show that $\rho > 0$.

(b) Show that $\mathbf{w}^T(t) w^* \geq \mathbf{w}^T (t-1)\mathbf{w}^* + \rho$, and conclude that $\mathbf{w}^T(t) \mathbf{w}^* \geq t\rho$. <br />
[__Hint__: Use induction]

(c) Show that $\parallel \mathbf{w}(t) \parallel^2 \leq \parallel \mathbf{w}(t-1) \parallel^2 + \parallel \mathbf{x}(t-1) \parallel^2$. <br />
[__Hint__: $y(t-1) \dot (\mathbf{w}^T (t-1) \mathbf{x} (t-1) \leq 0$ because $\mathbf{x}(t-1)$ was misclassified by $\mathbf{w}(t-1)$.]

(d) Show by induction that $\parallel \mathbf{w}(t) \parallel^2 \leq tR^2$, where $R = \max_{1 \leq n \leq N} \parallel \mathbf{x}_n \parallel$.

(e) Using (b) and (d), show that
    \begin{equation*}
        \frac{\mathbf{w}^T(t)}{\parallel \mathbf{w}(t) \parallel} \mathbf{w}^* \geq \sqrt{t} \frac{\rho}{R} \text{,}
    \end{equation*}
    and hence prove that
    \begin{equation*}
        t \leq \frac{R^2 \parallel \mathbf{w}^* \parallel^2}{\rho^2}
    \end{equation*}
[__Hint__: $\frac{\mathbf{w}^T(t)\mathbf{w}^*}{\parallel \mathbf{w}(t) \parallel \parallel \mathbf{w}^* \parallel} \leq 1$. Why?]

In practice, PLA converges more quickly than the bound $\frac{R^2 \parallel \mathbf{w}^* \parallel^2}{\rho^2}$ suggests. Nevertheless, because we do not know $\rho$ in advance, we can't determine the number of iterations to converge, which does pose a problem if the data is non-separable.
    


# <font color='Red'>Solution</font>

<b>(a)</b> 
Since $\mathbf{w}^*$ classifies every $\mathbf{x}_n$ correctly for $n=1,2,...N$, we have $$y_n = sign(\mathbf{w}^{*T} \mathbf{x}_n)$$  <br>
which is equivalent to $$y_n (\mathbf{w}^{*T}\mathbf{x}_n) >0$$ <br>
Since it is true for all $n$, it is also true for the minimum $$ \rho = \min_{1 \leq n \leq N} y_n(\mathbf{w}^{*T}\mathbf{x}_n) > 0$$


<b>(b)</b>
We know, 
\begin{aligned}
\mathbf{w}^T(t)\mathbf{w}^* &= [\mathbf{w}(t-1) + y(t-1) \mathbf{x}(t-1)]^T \mathbf{w}^*\\
&=[\mathbf{w}^T(t-1) + y(t-1) \mathbf{x}^T(t-1)] \mathbf{w}^* \\
&\geq \mathbf{w}^T(t-1)\mathbf{w}^* + \rho 
\end{aligned} <br>
Since $\rho = \min_{1 \leq n \leq N} y(t-1)\mathbf{x}^T(t-1)\mathbf{w}^{*}$ <br>
        Now let's use induction on the above equation: <br>
        <ol>
            <li> If $t = 0$,  $\mathbf{w}^T(0) \mathbf{w}^* \geq 0 $ [As $\mathbf{w}(0)=0$]</li>
            <li> Consider the proposition is true for $t$, then we have  $\mathbf{w}^T(t)\mathbf{w}^* \geq t \rho$</li>
            <li> Now for $t = t+1$,<br>
                \begin{aligned}
                \mathbf{w}^T(t+1)\mathbf{w}^* &= \mathbf{w}^T(t)\mathbf{w}^* +  y(t) \mathbf{x}^T(t) \mathbf{w}^*\\
                & \geq t \rho + \rho \\
                & \geq (t+1)\rho
                \end{aligned}<br>
                Therefore, $\mathbf{w}^T(t)\mathbf{w}^* \geq t \rho$ <b>[Proved]</b> 
             </li>
        </ol>
                
                
<b>(c)</b>
                
Now, we have,
\begin{aligned}
||\mathbf{w}(t)||^2 &= ||\mathbf{w}(t-1) + y(t-1)\mathbf{x}(t-1)||^2\\
&= ||\mathbf{w}(t-1)||^2 + ||y(t-1)\mathbf{x}(t-1)||^2 + 2y(t-1) \mathbf{w}^T(t-1)\mathbf{x}(t-1)\\
& \leq ||\mathbf{w}(t-1)||^2 + ||y(t-1)\mathbf{x}(t-1)||^2\\
& \leq ||\mathbf{w}(t-1)||^2 + ||\mathbf{x}(t-1)||^2\\
\end{aligned}<br>
where $\mathbf{x}(t-1)$ is misclassified by $\mathbf{w}(t-1)$. Therefore, $y(t-1) \mathbf{w}^T(t-1)\mathbf{x}(t-1) <0$, and $||y(t-1)||^2 = 1$<br>
    


<b> (d) </b>
Using induction we will prove $||\mathbf{w}(t)||^2 \leq tR^2$, where $R = \max_{1\leq n \leq N} ||\mathbf{x}_n||$
        <ol>
            <li> For $t=0$,  $||\mathbf{w}(0)||^2 \leq 0 \cdot R^2$,</li>
            <li> Consider the proposition is true for $t=t-1$, then we have  $||\mathbf{w}(t-1)||^2 \leq (t-1)R^2$ </li>
            <li> Now for $t$, using the equation in (c) we can write,<br>
                \begin{aligned}
                ||\mathbf{w}(t)||^2 & \leq ||\mathbf{w}(t-1)||^2 + ||\mathbf{x}(t-1)||^2\\
                & \leq  (t-1)R^2 + ||\mathbf{x}(t-1)||^2\\
                & \leq  (t-1)R^2 + R^2\\
                & = tR^2
                \end{aligned}<br>
                where $||\mathbf{x}(t-1)||^2 \leq R^2 = (\max_{1\leq n \leq N} ||\mathbf{x}_n||)^2 $</li>
        </ol>
        
<b>(e)</b>
Using (b) and (d) we get,
$$\begin{aligned}
\frac{\mathbf{w}^T(t)}{||\mathbf{w}(t)||} \mathbf{w}^*  & \geq t\rho \frac{1}{||\mathbf{w}(t)||}\\
&\geq t\rho \frac{1}{\sqrt{tR^2}}\\
& = \sqrt{t} \frac{\rho}{R}
\end{aligned}$$
This equation can be rewritten as 
$$\begin{aligned}
t  & \leq \frac{R^2}{\rho^2} \frac{(\mathbf{w}^T(t)\mathbf{w}^*)^2}{||\mathbf{w}(t)||^2}
\end{aligned}$$ <br>

Now, using Cauchy-Schwarz inequality ($|a \cdot b| = ||a|| \cdot ||b|| cos \theta \leq ||a|| \cdot ||b||$ as $cos \theta \leq 1$),<br>
$$ \frac{(\mathbf{w}^T(t)\mathbf{w}^*)^2}{||\mathbf{w}(t)||^2 ||\mathbf{w}^*||^2} \leq 1$$
So, we can write,
$$\begin{aligned}
t  & \leq \frac{R^2}{\rho^2} \frac{(\mathbf{w}^T(t)\mathbf{w}^*)^2}{||\mathbf{w}(t)||^2} \frac{||\mathbf{w}^*||^2}{||\mathbf{w}^*||^2}\\
& \leq \frac{R^2 ||\mathbf{w}^*||^2}{\rho^2}
\end{aligned}$$ <br>

# <font color='Red'>Solution</font>

```python
# Imports

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define a target function

def targetFunction(x, w):
    f = np.dot(w,x.T)
    return f
```

```python
# Define a sign function

def sign(f):
    return np.sign(f)
```

```python
# Define a data generator

def dataGenerator(dimension, num_samples, seed=5):
    np.random.seed(seed)

    # randomly generate data from uniform distribution
    if dimension == 2:
        # if there are only 2 dimension add column names to plot the results
        data_x = pd.DataFrame(np.random.uniform(-100,100,(num_samples, dimension)), columns=["x_1","x_2"])
    else:
        # if dimensions are greater than 2 then it is not possible to visualize
        data_x = pd.DataFrame(np.random.uniform(-100,100,(num_samples, dimension)))
    data_x.insert(0, "x_0", np.ones(num_samples))
    
    # randomly select weights
    weights = np.random.uniform(-1, 1, (1, dimension + 1))
    
    # generate the data labels
    f = targetFunction(data_x, weights)
    data_y = sign(f)[0]
    
    # dataframe consisting of data and labels
    data_df = pd.concat([data_x, pd.DataFrame(data_y, columns = ['y'])], axis=1)
    
    return data_df, data_x, data_y, weights
```

```python
# Function to plot decision boundary

def plotLine(slope, intercept, label, color):
    axes = plt.gca()
    x_vals = np.array(axes.get_xlim())
    y_vals = intercept + (slope * x_vals)
    plt.plot(x_vals, y_vals, label=label, c=color)
    plt.legend()
```

```python
# Function to plot the data

def plotData(data, weights):
    data_groups = data.groupby('y')
    for name, group in data_groups:
        plt.plot(group.x_1, group.x_2, marker='o', linestyle='', markersize=7, label=name)

    plt.legend()
    plotLine(-weights[0,1]/weights[0,2], -weights[0,0]/weights[0,2], label="f", color="red")
    plt.xlabel("x_1")
    plt.ylabel("x_2")
    plt.show()
```

```python
# Perceptron function to train on the data

def perceptron(x, y, method="deterministic"):
    
    x_arr = np.array(x)
    y_arr = np.array(y)
    weight_arr = np.zeros(x_arr.shape[1])

    # find misclassified samples
    y_pred = sign(targetFunction(x_arr, weight_arr))

    n_iter = 0
  
    while np.any(y_pred != y_arr):

        # get misclassified samples
        mis_ind = np.where(y_pred != y_arr)
        
        if method == "deterministic":
            index = mis_ind[0][0]
        else:
            index = np.random.choice(mis_ind[0])
        
        x_mis = x_arr[index]
        y_mis = y_arr[index]
        
        # update weights
        weight_arr += y_mis * x_mis
        
        # find misclassified samples
        y_pred = sign(targetFunction(x_arr, weight_arr))
        # update number of iterations
        n_iter += 1
        
    return weight_arr, n_iter
```

```python
# Function to plot results

def plotResults(data, orig_weights, appr_weights, n_iter):
    data_groups = data.groupby('y')
    for name, group in data_groups:
        plt.plot(group.x_1, group.x_2, marker='o', linestyle='', markersize=7, label=name)

    plt.legend()
    plotLine(-orig_weights[0,1]/orig_weights[0,2], -orig_weights[0,0]/orig_weights[0,2], label="f", color="red")
    plotLine(-appr_weights[1]/appr_weights[2], -appr_weights[0]/appr_weights[2], label="g", color="blue")

    plt.xlim(-100, 100)
    plt.ylim(-100, 100)
    plt.xlabel("x_1")
    plt.ylabel("x_2")
    plt.title(f'Perceptron need {n_iter} iterations to converge')
    plt.show()
```

# <font color='blue'>Exercise 3</font>
The perceptron learning algorithm works like this: In each iteration $t$, pick a random $(\mathbf{x}(t), y(t))$ and compute the 'signal' $s(t) = \mathbf{w}^T(t) \mathbf{x}(t)$. If $y(t) \cdot s(t) \leq 0$, update $\mathbf{w}$ by

\begin{equation*}
    \mathbf{w}(t+1) \gets \mathbf{w}(t) + y(t) \cdot \mathbf{x}(t) ,
\end{equation*}

One may argue that this algorithm does not take the 'closeness' between $s(t)$ and $y(t)$ into consideration. Let's look at another perceptron learning algorithm. In each iteration, pick a random $(\mathbf{x}(t), y(t))$ and compute $s(t)$. If $y(t) \cdot s(t) \leq 1$, update $\mathbf{w}$ by

\begin{equation*}
    \mathbf{w}(t+1) \gets \mathbf{w}(t) + \eta \cdot (y(t) - s(t)) \cdot \mathbf{x}(t) ,
\end{equation*}

where $\eta$ is a constant. That is, if $s(t)$ agrees with $y(t)$ well (their product is $>1$), the algorithm does nothing. On the other hand, if $s(t)$ is further from $y(t)$, the algorithm changes $\mathbf{w}(t)$ more. In this problem, you are asked to implement this algorithm and study its performance.

(a) Generate a training data set of size $100$ similar to that used in Exercise 2(a). Generate a test data set of size $10,000$ from the same process. To get $g$, run the algorithm above $\eta = 100$ on the training data set until a maximum of $1,000$ updates has been reached. Plot the training data set, the target function $f$, and the final hypothesis $g$ on the same figure. Report the error on the test set.

(b) Use the data set in (a) and redo everything with $\eta = 1$.

(c) Use the data set in (a) and redo everything with $\eta = 0.01$.

(d) Use the data set in (a) and redo everything with $\eta = 0.0001$.

(d) Compare the results that you get from (a) to (d).

The algorithm above is a  variant of the so-called Adaline (Adaptive Linear Neuron) algorithm for perceptron learning.

```python
# Adaline function to train on the data

def adaline(x, y, eta, max_iter, method="deterministic"):
    np.random.seed(40)
    x_arr = np.array(x)
    y_arr = np.array(y)
    weight_arr = np.random.uniform(low=-1, high=1, size=x_arr.shape[1])

    # find misclassified samples
    s = targetFunction(x_arr, weight_arr)

    prod = s * y_arr

    n_iter = 0
  
    while np.any(prod <= 1):

        # get misclassified samples
        mis_ind = np.where(prod <= 1)
        
        if method == "deterministic":
            index = mis_ind[0][0]
        else:
            index = np.random.choice(mis_ind[0])
        
        x_mis = x_arr[index]
        y_mis = y_arr[index]
        s_mis = s[index]

        # bounding to prevent overflow
        if s_mis < -100:
          s_mis = -100
        elif s_mis > 100:
          s_mis = 100
        
        # update weights
        weight_arr += eta * (y_mis - s_mis) * x_mis
        
        # find misclassified samples
        s = targetFunction(x_arr, weight_arr)
        prod = s * y_arr

        # update number of iterations
        n_iter += 1

        if n_iter >= max_iter:
          break
        
    return weight_arr, n_iter
```

```python
# Test the Adaline model

def test_model(test_x, test_y, approx_weight):

  s_test = targetFunction(np.array(test_x), approx_weight)
  prod_test = s_test * np.array(test_y)

  incorrect = len(np.where(prod_test <= 1)[0])

  print(f'Number of misclassified points {incorrect} | Accuracy {(len(test_x) - incorrect)/len(test_x)}')
```

```python
import random

random.seed(40)

# Generate 10100 total data samples with 2 dimensions
total_df, total_x, total_y, weights_total = dataGenerator(num_samples = 10100, dimension = 2, seed = 40)

all_ind = list(range(len(total_x)))
train_ind = random.sample(all_ind, 100)
test_ind = list(set(all_ind) - set(train_ind))

# Split the data in train and test
train_x = total_x.iloc[train_ind].reset_index(drop=True)
train_y = total_y[train_ind]
train_df = pd.concat([train_x, pd.DataFrame(train_y, columns = ['y'])], axis=1)

test_x = total_x.iloc[test_ind].reset_index(drop=True)
test_y = total_y[test_ind]
test_df = pd.concat([test_x, pd.DataFrame(test_y, columns = ['y'])], axis=1)
```

```python
# Execute Adaline
approx_weight_6, num_iter_6 = adaline(train_x, train_y, eta = 100, max_iter=1000)

# Plot the results from Adaline
plotResults(train_df, weights_total, approx_weight_6, num_iter_6)

# Report test error or accuracy
test_model(test_x, test_y, approx_weight_6)
```
